```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/Metadata_Country.csv')

```

```python
df1.head()
```

```python
df1.rename(columns={'TableName': 'Country_Name'}, inplace=True)
df1 = df1.drop('Unnamed: 5', axis=1)
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
# @title Region vs IncomeGroup

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(6, 6))
df_2dhist = pd.DataFrame({
    x_label: grp['IncomeGroup'].value_counts()
    for x_label, grp in df1.groupby('Region')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Region')
_ = plt.ylabel('IncomeGroup')
```

```python
# @title Region

from matplotlib import pyplot as plt
import seaborn as sns
df1.groupby('Region').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title IncomeGroup

from matplotlib import pyplot as plt
import seaborn as sns
df1.groupby('IncomeGroup').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
